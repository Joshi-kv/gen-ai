```python
from transformers import pipeline
```

```python
text = "Whether you're a student, a hobbyist, or a ML researcher, Colab has you covered.Colab is always free of charge to use, but as your computing needs grow there are paid options to meet them."
```

```python
pipe = pipeline("text-to-speech", model="facebook/mms-tts-eng", device="cuda")
```

```python
output = pipe(text)
```

```python
output
```

```python
from IPython.display import Audio
```

```python
Audio(output['audio'], rate=output['sampling_rate'])
```
