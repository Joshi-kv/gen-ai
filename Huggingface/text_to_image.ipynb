```python
from diffusers import StableDiffusionPipeline
import torch
from matplotlib import pyplot as plt
```

```python
model1 = "dreamlike-art/dreamlike-diffusion-1.0"
model2 = "stabilityai/stable-diffusion-xl-base-1.0"
```

```python
pipe = StableDiffusionPipeline.from_pretrained(model1, torch_dtype=torch.float16, use_auth_token=True)
pipe = pipe.to("cuda")
```

```python
prompt = "An astronaut riding a green horse"
```

```python
image = pipe(prompt)
```

```python
image.images[0]
```

```python
def generate_image(pipe, prompt, params):
  img = pipe(prompt, **params).images
  image_len = len(img)
  if image_len == 1:
    return img[0]
  else:
    fig, ax = plt.subplots(1, image_len, figsize=(20, 20))
    for i in range(image_len):
      ax[i].imshow(img[i])
      ax[i].axis("off")
```

```python
prompt = "Orange cat with guitar"
params = {'num_images_per_prompt': 2}
```

```python
generate_image(pipe, prompt, params)
```
